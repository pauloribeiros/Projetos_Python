```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```
bnb = pd.read_csv('/content/listings Rio de Janeiro.csv', low_memory=False)
bnb.head()
```

```
features = ['name']
bnb = bnb[features]
```

```
bnb = bnb.dropna(axis=0)
```

```
bnb.head()
```

```
bnb_name =" ".join(s for s in bnb['name'])
```

```
stopwords = set(STOPWORDS)
```

```
wordcloud_name = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(bnb_name)
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud_name , interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud_name);
```
